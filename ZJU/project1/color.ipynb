```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取原始黑白图像
original_image = cv2.imread('Scene.jpg', cv2.IMREAD_COLOR)

# 判断是否正确导入图像
if original_image is None:
    print("Error: Unable to load image.")
    exit()

cv2.imshow('Origin Image', original_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
red_channel = original_image[:, :, 0]  
green_channel = original_image[:, :, 1]  
blue_channel = original_image[:, :, 2]  
```

```python
# 创建一个2x3的子图布局
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# 显示红色通道图像
axes[0].imshow(red_channel)
axes[0].set_title('Red Channel')

# 显示绿色通道图像
axes[1].imshow(green_channel)
axes[1].set_title('Green Channel')

# 显示蓝色通道图像
axes[2].imshow(blue_channel)
axes[2].set_title('Blue Channel')

# 隐藏坐标轴
for ax in axes:
    ax.axis('off')

# 调整子图之间的间距
plt.tight_layout()

# 显示图像
plt.show()
```

```python
processed_image = cv2.merge([red_channel, green_channel, blue_channel])

cv2.imshow('Merged Image', processed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
